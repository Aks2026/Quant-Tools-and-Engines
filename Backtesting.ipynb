```python
from datetime import datetime, timedelta
from typing import Dict, List, Tuple, Optional
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

class PortfolioBacktester:
    """
    Accurate portfolio backtesting engine with proper weight drift and churn-based costs
    Uses typical price (High + Low + Close) / 3 instead of close price
    Fixed to handle duplicate symbols properly
    """
    
    def __init__(self, port_df: pd.DataFrame, price_df: pd.DataFrame, 
                 initial_capital: float = 1000000, 
                 churn_cost: float = 0.015, fixed_cost: float = 0.01,
                 use_typical_price: bool = True):
        """
        Initialize the backtester
        
        Args:
            port_df: DataFrame with columns ['Date', 'Symbol', 'Weight']
            price_df: DataFrame with columns ['Date', 'Symbol', 'Open', 'High', 'Low', 'Close', ...]
            initial_capital: Starting capital amount
            churn_cost: Cost per 1% churn (0.015 = 1.5%)
            fixed_cost: Fixed cost per rebalance (0.01 = 1%)
            use_typical_price: If True, uses (High+Low+Close)/3, else uses Close
        """
        self.port_df = port_df.copy()
        self.price_df = price_df.copy()
        self.initial_capital = initial_capital
        self.churn_cost = churn_cost
        self.fixed_cost = fixed_cost
        self.use_typical_price = use_typical_price
        
        # Prepare data
        self._prepare_data()
        
    def _prepare_data(self):
        """Prepare and clean the input data"""
        # Convert dates
        self.port_df['Date'] = pd.to_datetime(self.port_df['Date'])
        self.price_df['Date'] = pd.to_datetime(self.price_df['Date'])
        
        # Calculate price to use
        if self.use_typical_price:
            # Calculate typical price = (High + Low + Close) / 3
            self.price_df['Price'] = (self.price_df['High'] + self.price_df['Low'] + self.price_df['Close']) / 3
            price_column = 'Price'
            print("Using typical price: (High + Low + Close) / 3")
        else:
            # Use close price
            self.price_df['Price'] = self.price_df['Close']
            price_column = 'Close'
            print("Using close price")
        
        # Sort data
        self.port_df = self.port_df.sort_values(['Date', 'Symbol'])
        self.price_df = self.price_df.sort_values(['Date', 'Symbol'])
        
        # Check for and handle duplicate symbols in portfolio data
        duplicates = self.port_df.groupby(['Date', 'Symbol']).size()
        if (duplicates > 1).any():
            print("Warning: Found duplicate symbols in portfolio data. Aggregating weights...")
            # Aggregate duplicate symbols by summing their weights
            self.port_df = self.port_df.groupby(['Date', 'Symbol'])['Weight'].sum().reset_index()
            print(f"Aggregated {(duplicates > 1).sum()} duplicate entries")
        
        # Check for and handle duplicate symbols in price data
        price_duplicates = self.price_df.groupby(['Date', 'Symbol']).size()
        if (price_duplicates > 1).any():
            print("Warning: Found duplicate symbols in price data. Taking first occurrence...")
            # Take first occurrence for duplicate prices
            self.price_df = self.price_df.drop_duplicates(subset=['Date', 'Symbol'], keep='first')
            print(f"Removed {(price_duplicates > 1).sum()} duplicate price entries")
        
        # Create price pivot table
        self.price_pivot = self.price_df.pivot_table(
            index='Date', columns='Symbol', values='Price', aggfunc='first'
        )
        
        # Forward fill missing prices
        self.price_pivot = self.price_pivot.fillna(method='ffill')
        
        # Get unique portfolio dates
        self.portfolio_dates = sorted(self.port_df['Date'].unique())
        
        # Get all trading dates starting from first portfolio date
        first_portfolio_date = self.portfolio_dates[0]
        self.trading_dates = [d for d in sorted(self.price_pivot.index) if d >= first_portfolio_date]
        
        print(f"Data Summary:")
        print(f"  Portfolio dates: {len(self.portfolio_dates)}")
        print(f"  First portfolio date: {first_portfolio_date.date()}")
        print(f"  Trading dates (from first portfolio): {len(self.trading_dates)}")
        print(f"  Date range: {self.trading_dates[0].date()} to {self.trading_dates[-1].date()}")
        
        # Show sample of price calculation
        sample_data = self.price_df.head(3)[['Date', 'Symbol', 'High', 'Low', 'Close', 'Price']]
        print(f"\nSample price calculation:")
        print(sample_data.round(2))
        
    def _get_portfolio_for_date(self, date: pd.Timestamp) -> pd.DataFrame:
        """Get the active portfolio for a given date"""
        # Find the most recent portfolio date <= given date
        applicable_date = None
        for port_date in self.portfolio_dates:
            if port_date <= date:
                applicable_date = port_date
            else:
                break
        
        if applicable_date is None:
            return pd.DataFrame()
        
        portfolio = self.port_df[self.port_df['Date'] == applicable_date].copy()
        
        # Additional safety check: remove any remaining duplicates
        if portfolio['Symbol'].duplicated().any():
            print(f"Warning: Found duplicates in portfolio for {date.date()}, aggregating...")
            portfolio = portfolio.groupby('Symbol')['Weight'].sum().reset_index()
            portfolio['Date'] = applicable_date
        
        return portfolio
    
    def _create_weight_series(self, portfolio_df: pd.DataFrame) -> pd.Series:
        """Safely create a weight series from portfolio DataFrame, handling duplicates"""
        if portfolio_df.empty:
            return pd.Series(dtype=float)
        
        # Group by symbol and sum weights (handles any remaining duplicates)
        weight_series = portfolio_df.groupby('Symbol')['Weight'].sum()
        
        # Normalize weights to sum to 1
        if weight_series.sum() > 0:
            weight_series = weight_series / weight_series.sum()
        
        return weight_series
    
    def _calculate_churn(self, old_weights: pd.Series, new_weights: pd.Series) -> float:
        """Calculate portfolio churn percentage"""
        if old_weights.empty:
            return 1.0  # 100% churn for initial allocation
        
        # Ensure no duplicate indices
        if old_weights.index.duplicated().any():
            print("Warning: Duplicate symbols in old_weights, aggregating...")
            old_weights = old_weights.groupby(old_weights.index).sum()
        
        if new_weights.index.duplicated().any():
            print("Warning: Duplicate symbols in new_weights, aggregating...")
            new_weights = new_weights.groupby(new_weights.index).sum()
        
        # Align the two weight series
        all_symbols = old_weights.index.union(new_weights.index)
        old_aligned = old_weights.reindex(all_symbols, fill_value=0)
        new_aligned = new_weights.reindex(all_symbols, fill_value=0)
        
        # Calculate weight changes
        weight_changes = new_aligned - old_aligned
        
        # Churn = sum of positive changes (buys) = sum of absolute negative changes (sells)
        # We take the sum of positive changes (money being invested in new/increased positions)
        churn = weight_changes[weight_changes > 0].sum()
        
        # Churn cannot exceed 100%
        return min(churn, 1.0)
    
    def _show_weight_drift_sample(self, start_date: pd.Timestamp, days: int = 7) -> pd.DataFrame:
        """Show detailed weight drift calculation for sample period"""
        
        # Find start date index
        start_idx = None
        for i, date in enumerate(self.trading_dates):
            if date >= start_date:
                start_idx = i
                break
        
        if start_idx is None:
            return pd.DataFrame()
        
        # Get initial portfolio
        initial_portfolio = self._get_portfolio_for_date(start_date)
        if initial_portfolio.empty:
            return pd.DataFrame()
        
        # Initialize weights using safe method
        weights = self._create_weight_series(initial_portfolio)
        
        nav = self.initial_capital
        sample_data = []
        
        end_idx = min(start_idx + days, len(self.trading_dates))
        
        for i in range(start_idx, end_idx):
            date = self.trading_dates[i]
            current_prices = self.price_pivot.loc[date].dropna()
            
            if i > start_idx:
                prev_date = self.trading_dates[i-1]
                prev_prices = self.price_pivot.loc[prev_date].dropna()
                
                # Calculate stock returns
                common_symbols = weights.index.intersection(current_prices.index).intersection(prev_prices.index)
                
                if len(common_symbols) > 0:
                    stock_returns = (current_prices[common_symbols] - prev_prices[common_symbols]) / prev_prices[common_symbols]
                    stock_returns = stock_returns.fillna(0)
                    
                    # Calculate portfolio return using current weights
                    portfolio_return = (weights[common_symbols] * stock_returns).sum()
                    
                    # Update NAV
                    nav = nav * (1 + portfolio_return)
                    
                    # Update weights due to price changes (weight drift)
                    if portfolio_return != -1:
                        new_weights = weights[common_symbols] * (1 + stock_returns) / (1 + portfolio_return)
                        # Normalize weights to sum to 1 (in case some stocks are missing)
                        if new_weights.sum() > 0:
                            weights = new_weights / new_weights.sum()
                        else:
                            weights = new_weights
            
            # Store sample data
            sample_data.append({
                'Date': date,
                'NAV': nav,
                'Daily_Return': (nav / self.initial_capital - 1) if i == start_idx else portfolio_return,
                'Max_Weight': weights.max() if not weights.empty else 0,
                'Min_Weight': weights.min() if not weights.empty else 0,
                'Num_Holdings': len(weights),
                'Weight_Sum': weights.sum() if not weights.empty else 0
            })
        
        return pd.DataFrame(sample_data)
    
    def run_backtest(self, show_sample: bool = True) -> pd.DataFrame:
        """Run the backtest with proper weight drift and churn costs"""
        print(f"\nStarting backtest using {'typical price' if self.use_typical_price else 'close price'}...")
        
        # Show sample weight drift calculation
        if show_sample:
            first_date = self.trading_dates[0]
            sample_df = self._show_weight_drift_sample(first_date, days=7)
            print(f"\nSample Weight Drift Calculation (first 7 days from {first_date.date()}):")
            print(sample_df.round(6))
        
        results = []
        current_weights = pd.Series(dtype=float)
        nav = self.initial_capital
        is_first_allocation = True
        
        for i, date in enumerate(self.trading_dates):
            if i % 1000 == 0:
                print(f"Processing {date.date()} ({i+1}/{len(self.trading_dates)})")
            
            # Get current and previous prices
            current_prices = self.price_pivot.loc[date].dropna()
            
            if i > 0:
                prev_prices = self.price_pivot.loc[self.trading_dates[i-1]].dropna()
            else:
                prev_prices = current_prices.copy()
            
            # Calculate daily returns and weight drift (before any rebalancing)
            daily_portfolio_return = 0
            if not current_weights.empty and i > 0:
                # Calculate stock returns
                common_symbols = current_weights.index.intersection(current_prices.index).intersection(prev_prices.index)
                
                if len(common_symbols) > 0:
                    stock_returns = (current_prices[common_symbols] - prev_prices[common_symbols]) / prev_prices[common_symbols]
                    stock_returns = stock_returns.fillna(0)
                    
                    # Calculate portfolio return using current (drifted) weights
                    daily_portfolio_return = (current_weights[common_symbols] * stock_returns).sum()
                    
                    # Update NAV based on portfolio return
                    nav = nav * (1 + daily_portfolio_return)
                    
                    # Update weights due to price changes (weight drift)
                    if daily_portfolio_return != -1:
                        new_weights = current_weights[common_symbols] * (1 + stock_returns) / (1 + daily_portfolio_return)
                        current_weights = new_weights
            
            # Check for rebalancing
            should_rebalance = False
            if date in self.portfolio_dates:
                should_rebalance = True
                
                # Get target portfolio
                target_portfolio = self._get_portfolio_for_date(date)
                
                if not target_portfolio.empty:
                    # Get target weights using safe method
                    target_weights = self._create_weight_series(target_portfolio)
                    
                    if is_first_allocation:
                        print(f"\nInitial allocation on {date.date()}")
                        churn = 1.0  # 100% churn for initial allocation
                        transaction_cost = 0.0  # No cost for initial allocation
                        print(f"  Initial capital: ₹{nav:,.0f}")
                        print(f"  Transaction cost: ₹0 (no cost on initial allocation)")
                    else:
                        print(f"\nRebalancing on {date.date()}")
                        churn = self._calculate_churn(current_weights, target_weights)
                        transaction_cost = churn * self.churn_cost + self.fixed_cost
                        print(f"  Pre-rebalance NAV: ₹{nav:,.0f}")
                        print(f"  Portfolio churn: {churn:.2%}")
                        print(f"  Churn cost ({churn:.1%} × 1.5%): ₹{nav * churn * self.churn_cost:,.0f}")
                        print(f"  Fixed cost (1%): ₹{nav * self.fixed_cost:,.0f}")
                        print(f"  Total cost: ₹{nav * transaction_cost:,.0f}")
                    
                    # Apply transaction costs
                    nav = nav * (1 - transaction_cost)
                    
                    # Set new weights
                    current_weights = target_weights.copy()
                    
                    print(f"  Post-cost NAV: ₹{nav:,.0f}")
                    print(f"  New portfolio: {len(current_weights)} stocks")
                    
                    is_first_allocation = False
            
            # Store results
            results.append({
                'Date': date,
                'NAV': nav,
                'Daily_Return': daily_portfolio_return,
                'Cumulative_Return': (nav - self.initial_capital) / self.initial_capital,
                'Holdings_Count': len(current_weights),
                'Weight_Concentration': current_weights.max() if not current_weights.empty else 0,
                'Is_Rebalance_Day': should_rebalance,
                'Weight_Sum': current_weights.sum() if not current_weights.empty else 0
            })
        
        # Convert to DataFrame
        results_df = pd.DataFrame(results)
        
        print(f"\nBacktest completed!")
        print(f"Final NAV: ₹{results_df['NAV'].iloc[-1]:,.0f}")
        print(f"Total return: {results_df['Cumulative_Return'].iloc[-1]:.2%}")
        print(f"Number of rebalances: {results_df['Is_Rebalance_Day'].sum()}")
        
        return results_df
    
    def analyze_missing_stocks(self, results_df: pd.DataFrame) -> pd.DataFrame:
        """Analyze periods where portfolio weights sum to less than 1.0"""
        
        weight_issues = results_df[results_df['Weight_Sum'] < 0.99].copy()
        
        if len(weight_issues) > 0:
            print(f"\nFound {len(weight_issues)} days where weights sum < 99%:")
            print(weight_issues[['Date', 'Weight_Sum', 'Holdings_Count']].head(10))
            
            # Group by consecutive periods
            weight_issues['Date_Diff'] = weight_issues['Date'].diff().dt.days
            weight_issues['Period_Group'] = (weight_issues['Date_Diff'] > 1).cumsum()
            
            periods = weight_issues.groupby('Period_Group').agg({
                'Date': ['min', 'max'],
                'Weight_Sum': ['min', 'mean'],
                'Holdings_Count': 'mean'
            }).round(4)
            
            periods.columns = ['Start_Date', 'End_Date', 'Min_Weight_Sum', 'Avg_Weight_Sum', 'Avg_Holdings']
            print(f"\nPeriods with weight sum issues:")
            print(periods.head())
            
        return weight_issues
    
    def get_performance_summary(self, results_df: pd.DataFrame) -> Dict:
        """Calculate performance metrics"""
        
        # Time period
        start_date = results_df['Date'].iloc[0]
        end_date = results_df['Date'].iloc[-1]
        total_days = (end_date - start_date).days
        years = total_days / 365.25
        
        # Returns
        total_return = results_df['Cumulative_Return'].iloc[-1]
        annualized_return = (1 + total_return) ** (1/years) - 1 if years > 0 else total_return
        
        # Volatility
        daily_returns = results_df['Daily_Return'].dropna()
        volatility = daily_returns.std() * np.sqrt(252)
        
        # Sharpe ratio
        sharpe_ratio = annualized_return / volatility if volatility > 0 else 0
        
        # Maximum drawdown
        cumulative_returns = 1 + results_df['Cumulative_Return']
        running_max = cumulative_returns.expanding().max()
        drawdown = (cumulative_returns - running_max) / running_max
        max_drawdown = drawdown.min()
        
        # Win rate
        positive_days = (daily_returns > 0).sum()
        win_rate = positive_days / len(daily_returns) if len(daily_returns) > 0 else 0
        
        return {
            'Start_Date': start_date.date(),
            'End_Date': end_date.date(),
            'Years': years,
            'Total_Return': total_return,
            'Annualized_Return': annualized_return,
            'Volatility': volatility,
            'Sharpe_Ratio': sharpe_ratio,
            'Max_Drawdown': max_drawdown,
            'Win_Rate': win_rate,
            'Number_of_Rebalances': results_df['Is_Rebalance_Day'].sum(),
            'Final_NAV': results_df['NAV'].iloc[-1],
            'Initial_Capital': self.initial_capital,
            'Price_Type': 'Typical Price' if self.use_typical_price else 'Close Price'
        }
```

```python
# Ensure datetime format
results['Date'] = pd.to_datetime(results['Date'])
results = results.sort_values('Date').set_index('Date')

# Rebase NAV from the first available date (2006-06-19)
rebased_series = results['NAV']['2006-06-19':].copy()
base_value = rebased_series.iloc[0]
rebased_series = (rebased_series / base_value) * 100
rebased_series = rebased_series.sort_index()


# Daily returns
daily_returns = rebased_series.pct_change().dropna()

# Overall metrics
start_date = rebased_series.index[0]
end_date = rebased_series.index[-1]
days = (end_date - start_date).days
years = days / 365.25

cagr = (rebased_series.iloc[-1] / rebased_series.iloc[0]) ** (1 / years) - 1
rolling_max = rebased_series.cummax()
drawdown = rebased_series / rolling_max - 1
mdd = drawdown.min()
volatility = daily_returns.std() * np.sqrt(252)

# Calendar Year Returns
cy_returns = rebased_series.resample('Y').last().pct_change().dropna()
cy_returns.index = cy_returns.index.year
cy_df = pd.DataFrame({'CY Return': cy_returns})

# Financial Year Returns (Apr–Mar)
fy_nav = rebased_series.copy()
fy_nav.index = fy_nav.index.to_period('M')
fy_ends = fy_nav.groupby(fy_nav.index.to_timestamp().to_period("A-MAR")).last()
fy_returns = fy_ends.pct_change().dropna()
fy_returns.index = fy_returns.index.year + 1
fy_df = pd.DataFrame({'FY Return': fy_returns})

# Monthly rolling returns
monthly_periods = {
    '1M Return': 21,
    '2M Return': 42,
    '3M Return': 63,
    '6M Return': 126,
    '12M Return': 252,
    '24M Return': 504
}
monthly_data = []
for label, days_back in monthly_periods.items():
    if len(rebased_series) > days_back:
        ret = rebased_series.iloc[-1] / rebased_series.iloc[-days_back] - 1
        monthly_data.append({'Metric': label, 'Value': ret, 'Period': 'Latest'})

# Summary metrics
summary_df = pd.DataFrame({
    'Metric': ['CAGR', 'Max Drawdown', 'Annualized Volatility'],
    'Value': [cagr, mdd, volatility],
    'Period': 'Overall'
})

# Format Calendar Year Returns
cy_df_formatted = cy_df.rename_axis('Year').reset_index()
cy_df_formatted['Metric'] = 'CY Return'
cy_df_formatted = cy_df_formatted.rename(columns={'Year': 'Period', 'CY Return': 'Value'})

# Format Financial Year Returns
fy_df_formatted = fy_df.rename_axis('Year').reset_index()
fy_df_formatted['Metric'] = 'FY Return'
fy_df_formatted = fy_df_formatted.rename(columns={'Year': 'Period', 'FY Return': 'Value'})

# Combine everything
final_df = pd.concat([
    summary_df,
    pd.DataFrame(monthly_data),
    cy_df_formatted[['Metric', 'Period', 'Value']],
    fy_df_formatted[['Metric', 'Period', 'Value']]
], ignore_index=True)

# Format as percentage
final_df['Value'] = final_df['Value'].apply(lambda x: f"{x:.2%}")
final_df = final_df[['Metric', 'Period', 'Value']]

# Display final output/
final_df
```
